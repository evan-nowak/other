```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from jupyterthemes import jtplot
jtplot.style()
```

```python
x0, y0 = 0, 0
x1, y1 = 2, 4
```

```python
d = 1
```

```python
m = (y1 - y0) / (x1 - x0)
m
```

```python
def line_eq(x0, y0, m, x):
    return (x0 - x) / m + y0
```

```python
xs = np.linspace(-1, 1)
ys = line_eq(x0, y0, m, xs)
```

```python
plt.figure(figsize=(13, 7))
plt.plot(xs, ys, c='#987987')
plt.scatter([x0], [y0], s=200, c='#789789')
```
