```python
#!/usr/bin/env python3

import re
import os

fs = os.listdir()
fs = [i for i in fs if i.startswith('2')]

min_date = '"{0}"'.format(min(fs))
max_date = '"{0}"'.format(max(fs))

for f in fs:
    print(f)
    
    os.chdir(f)
    
    os.system('sudo rm alerts.html')
    os.system('sudo rm aoc_weekly.html')
    os.system('sudo rm offairport.html')
    
    # COVER
    
    with open('cover.html') as infile:
        s = infile.read()
        
    old = ';">[\d-]{1,3}%<'
    new = ';">--%<'

    s = re.sub(old, new, s)
    
    old = '"ops-value">[-\d,\.]+</div'
    new = '"ops-value">--</div'

    s = re.sub(old, new, s)
    
    old = '/dashboard/'
    new = '/sample/'
    
    s = re.sub(old, new, s)
    
    old = 'maxDate.yyyymmdd\(\)'
    new = max_date
    
    s = re.sub(old, new, s)

    old = 'minDateCov'
    new = min_date
    
    s = re.sub(old, new, s)
    
    old = '<div class="pa-alerts">.*?</td>'
    new = '<div class="pa-alerts"></div></td>'

    s = re.sub(old, new, s, flags=re.DOTALL)
    
    with open('cover.html', 'w') as outfile:
        outfile.write(s)
        
    # MTA
    
    os.chdir('mtabus')
    fns = os.listdir()
    
    for fn in fns:
        with open(fn) as infile:
            s = infile.read()

        old = '/dashboard/'
        new = '/sample/'

        s = re.sub(old, new, s)
    
        old = 'maxDate.yyyymmdd\(\)'
        new = max_date

        s = re.sub(old, new, s)

        old = 'minDateMTA'
        new = min_date

        s = re.sub(old, new, s)

        with open(fn, 'w') as outfile:
            outfile.write(s)
    
    os.chdir('..')
        
    # SHUTTLE
    
    os.chdir('shuttle')
    fns = os.listdir()
    
    for fn in fns:
        with open(fn) as infile:
            s = infile.read()

        old = '/dashboard/'
        new = '/sample/'

        s = re.sub(old, new, s)
    
        old = 'maxDate.yyyymmdd\(\)'
        new = max_date

        s = re.sub(old, new, s)

        old = 'minDateShuttle'
        new = min_date

        s = re.sub(old, new, s)

        with open(fn, 'w') as outfile:
            outfile.write(s)
    
    os.chdir('..')
        
    # WAZE
    
    os.chdir('waze')
    ffs = os.listdir()
    ffs = [i for i in ffs if i.startswith('l')]
    
    for ff in ffs:
        os.chdir(ff)
        fns = os.listdir()
        for fn in fns:
            with open(fn) as infile:
                s = infile.read()

            old = '/dashboard/'
            new = '/sample/'

            s = re.sub(old, new, s)
    
            old = 'maxDate.yyyymmdd\(\)'
            new = max_date

            s = re.sub(old, new, s)
    
            old = 'minDate'
            new = min_date

            s = re.sub(old, new, s)

            with open(fn, 'w') as outfile:
                outfile.write(s)
                
        os.chdir('..')
    
    # WAZE SPEEDS
    
    os.chdir('speeds')
    ffs = os.listdir()
    
    for ff in ffs:
        os.chdir(ff)
        fns = os.listdir()
        for fn in fns:
            with open(fn) as infile:
                s = infile.read()

            old = '/dashboard/'
            new = '/sample/'

            s = re.sub(old, new, s)
    
            old = 'maxDate.yyyymmdd\(\)'
            new = max_date

            s = re.sub(old, new, s)

            old = 'minDate'
            new = min_date

            s = re.sub(old, new, s)

            with open(fn, 'w') as outfile:
                outfile.write(s)
                
        os.chdir('..')
        
    os.chdir('..')
    
    os.chdir('..')
    
    os.chdir('..')
```
