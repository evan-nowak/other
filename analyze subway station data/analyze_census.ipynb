```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
df = pd.read_excel('data/census.xlsx')

print(df.shape)
df.head(2)
```

```python
df['boro_ct2010'] = df.id.str[-7:]

cols = ['median_income', 'employed_total', 'employed_service', 'employed_production', 'poverty']

for i in cols:
    df[i] = pd.to_numeric(df[i], errors='coerce')
```

```python
df['percent_service'] = df.employed_service / df.employed_total
df['percent_production'] = df.employed_production / df.employed_total
```
